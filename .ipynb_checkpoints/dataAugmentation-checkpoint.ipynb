# Data Augmentation for Pokemon Dataset

```python
import os
import tensorflow as tf
import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline 
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    #plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(width_shift_range=0.1, height_shift_range=0.1, shear_range=0.15, zoom_range=0.1, horizontal_flip=True)
```

```python
example = 'Pokemon_Dataset/abra.png'
```

```python
exampleImage = np.expand_dims(plt.imread(example),0)
plt.imshow(exampleImage[0])
```

```python
augmentationSample = gen.flow(exampleImage)
```

```python
print(augmentationSample)
```

```python
aug_images = [next(augmentationSample)[0].astype("float64") for i in range(10)]
```
